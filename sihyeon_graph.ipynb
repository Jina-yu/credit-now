```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.DataFrame([[0.8430, 0.8430, 0.8150, 0.8115, 0.8224],
                   [0.6592, 0.6676, 0.6592, 0.6918, 0.6915]], columns=['1 hidden', '2 hidden', '1 hidden by keras tuner', '5 hidden with keras tuner', '13 hidden with keras tuner hyperband'], index=['loss (cross-entropy)', 'test accuracy'])
```

```python
df = pd.DataFrame([['1 hidden', '2 hidden', '1 hidden by keras tuner', '5 hidden with keras tuner', '13 hidden with keras tuner hyperband', 'auto-keras'],
                   [0.6592, 0.6676, 0.6921, 0.6918, 0.6915, 0.63]])

#    [0.8430, 0.8430, 0.8150, 0.8115, 0.8224],
```

```python
df = df.transpose()
```

```python
df.columns = ['model', 'test_accuracy']
```

```python
df
```

```python
plt.figure(figsize=(30, 15))
sns.set(font_scale=1.7)
p = sns.barplot(data=df, 
            x = 'model',
            y = 'test_accuracy')
p.set_xlabel("Model", fontsize=25)
p.set_ylabel('Test Accuracy', fontsize=25)
p.set_yticks([0.5, 1])
```

```python
df_500 = pd.DataFrame([['Adam', 'Adagrad', 'RMSprop'],
                       [0.6752, 0.6881, 0.6745],
                   [0.6752, 0.6884, 0.6714]])

#    [0.8430, 0.8430, 0.8150, 0.8115, 0.8224],
```

```python
df_500 = df_500.transpose()
```

```python
df_500
```

```python
df_500.columns = ['Optimizer','Test_Accuracy_100', 'Test_Accuracy_500']
```

```python
fig = plt.figure(figsize=(20, 10))

area1 = fig.add_subplot(1, 2, 1)
area2 = fig.add_subplot(1 ,2, 2)

sns.set(font_scale=1.7)
p = sns.barplot(ax = area1, data=df_500, 
            x = 'Optimizer',
            y = 'Test_Accuracy_100')
p.set_xlabel("Model", fontsize=25)
p.set_ylabel('Test Accuracy (epoch = 100)', fontsize=25)
p.set_yticks([0.5, 1])
# p.bar_label(p.containers[0], label_type='edge')

sns.set(font_scale=1.7)
p = sns.barplot(ax = area2, data=df_500, 
            x = 'Optimizer',
            y = 'Test_Accuracy_500')
p.set_xlabel("Model", fontsize=25)
p.set_ylabel('Test Accuracy (epoch = 500)', fontsize=25)
p.set_yticks([0.5, 1])
# p.bar_label(p.containers[0], label_type='edge')
```
